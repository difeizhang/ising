## 1.1 Energy Data: Error vs. Number of Data

```python
def plot_error_vs_numsamples(data, ax, label):
    ax.plot(data['numsamples'], data['mean'], label=label)
    ax.fill_between(data['numsamples'], data['mean'] - data['std'], data['mean'] + data['std'], alpha=0.3)
```

```python
data_error_vs_numsample_histogram_SL = np.load('./overleaf/e/error_vs_numsample_histogram_SL.npz')
data_error_vs_numsample_histogram_LBC = np.load('./overleaf/e/error_vs_numsample_histogram_LBC.npz')
data_error_vs_numsample_histogram_PBM = np.load('./overleaf/e/error_vs_numsample_histogram_PBM.npz')
data_error_vs_numsample_discriminative_SL = np.load('./overleaf/e/error_vs_numsample_discriminative_SL.npz')
data_error_vs_numsample_discriminative_LBC = np.load('./overleaf/e/error_vs_numsample_discriminative_LBC.npz')
data_error_vs_numsample_discriminative_PBM = np.load('./overleaf/e/error_vs_numsample_discriminative_PBM.npz')
data_error_vs_numsample_generative_SL = np.load('./overleaf/e/error_vs_numsample_generative_SL.npz')
data_error_vs_numsample_generative_LBC = np.load('./overleaf/e/error_vs_numsample_generative_LBC.npz')
data_error_vs_numsample_generative_PBM = np.load('./overleaf/e/error_vs_numsample_generative_PBM.npz')
```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 4))
plot_error_vs_numsamples(data_error_vs_numsample_histogram_SL, axs[0], 'histogram')
plot_error_vs_numsamples(data_error_vs_numsample_discriminative_SL, axs[0], 'MLP')
plot_error_vs_numsamples(data_error_vs_numsample_generative_SL, axs[0], 'PixelCNN')
plot_error_vs_numsamples(data_error_vs_numsample_histogram_LBC, axs[1], 'histogram')
plot_error_vs_numsamples(data_error_vs_numsample_discriminative_LBC, axs[1], 'MLP')
plot_error_vs_numsamples(data_error_vs_numsample_generative_LBC, axs[1], 'PixelCNN')
plot_error_vs_numsamples(data_error_vs_numsample_histogram_PBM, axs[2], 'histogram')
plot_error_vs_numsamples(data_error_vs_numsample_discriminative_PBM, axs[2], 'MLP')
plot_error_vs_numsamples(data_error_vs_numsample_generative_PBM, axs[2], 'PixelCNN')

titles = ['SL', 'LBC', 'PBM']
for ax, title in zip(axs, titles):
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel('Number of samples')
    ax.set_ylabel('Error')
    ax.set_title(title)
    ax.legend()
plt.tight_layout()
plt.show()
```

## 2.1 Energy Data: Error vs. Training Time
### a. Without Date Generation Time

```python
def plot_error_vs_time(data, ax, label, mode='normal'):
    if mode == 'normal':
        ax.plot(data['time'], data['mean'], label=label)
        ax.fill_between(data['time'], data['mean'] - data['std'], data['mean'] + data['std'], alpha=0.3)
    elif mode == 'envelope':
        ax.plot(data['time'], data['error_env'], label=label)
    elif mode == 'scatter':
        ax.scatter(data['time_scatters'], data['error_scatters'], label=label, s=1, alpha=0.5)
    else:
        raise ValueError('Invalid mode')
```

```python
data_error_vs_time_histogram_SL = np.load('./overleaf/e/error_vs_time_histogram_SL.npz')
data_error_vs_time_histogram_LBC = np.load('./overleaf/e/error_vs_time_histogram_LBC.npz')
data_error_vs_time_histogram_PBM = np.load('./overleaf/e/error_vs_time_histogram_PBM.npz')
data_error_vs_time_discriminative_SL = np.load('./overleaf/e/error_vs_time_discriminative_SL.npz')
data_error_vs_time_discriminative_LBC = np.load('./overleaf/e/error_vs_time_discriminative_LBC.npz')
data_error_vs_time_discriminative_PBM = np.load('./overleaf/e/error_vs_time_discriminative_PBM.npz')
data_error_vs_time_generative_SL = np.load('./overleaf/e/error_vs_time_generative_SL.npz')
data_error_vs_time_generative_LBC = np.load('./overleaf/e/error_vs_time_generative_LBC.npz')
data_error_vs_time_generative_PBM = np.load('./overleaf/e/error_vs_time_generative_PBM.npz')
```
